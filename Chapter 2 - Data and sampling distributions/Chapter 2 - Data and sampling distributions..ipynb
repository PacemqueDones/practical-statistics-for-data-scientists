```python
import numpy as np
from scipy import stats

import seaborn as sns
import matplotlib.pyplot as plt

from prettytable import PrettyTable

from fractions import Fraction as F
```

```python
qtd = 100000

n=8; p = F(1,2)

binomial = np.random.binomial(n,p,qtd)

fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(12, 12))

sns.countplot(x=binomial, ax = axes[0])

bin1000 = np.random.choice(binomial,size=1000)

sns.histplot(bin1000, ax= axes[1], bins=n,binrange=(1,n+1))

mean5 = [np.mean(lista) for lista in [ np.random.choice(binomial,size=5) for i in range(1000)]]

sns.histplot(mean5, ax= axes[2], bins=n, binwidth=1,binrange=(1,n+1))

mean20 = [np.mean(lista) for lista in [ np.random.choice(binomial,size=20) for i in range(1000)]]

sns.histplot(mean20, ax= axes[3], bins=n, binwidth=1,binrange=(1,n+1))

table = PrettyTable()

table.field_names = [
    'Var da Distri',
    'Var da Distri amstl medias-size=5',
    'Std da Distri',
    'Std da Distri amstl-médias-size=5'
]

table.add_row(
    [float(n*p*(1-p)), 
     np.var(mean5)*5,
     np.sqrt(float(n*p*(1-p))),
     np.std(mean5)*np.sqrt(5)
     
    ]
)

table.add_row([
        'Var da Distri',
        'Var da Distri amstl médias-size=20',
        'Std da Distri',
        'Std da Distri amstl médias=size=20'
])

table.add_row(
    [float(n*p*(1-p)), 
     np.var(mean20)*20,
     np.sqrt(float(n*p*(1-p))),
     np.std(mean20)*np.sqrt(20)
    ]
)

table.hrules =True

print(table)
```

```python
qtd = 100000

mi=5; sig = 10

normal = np.random.normal(loc=mi,scale=sig,size=qtd)

fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(12, 12))

sns.histplot(x=normal, ax = axes[0])

normal1000 = np.random.choice(normal,size=1000)

sns.histplot(normal1000, ax= axes[1], binrange=(mi-3*sig,mi+3*sig))

mean5 = [np.mean(lista) for lista in [ np.random.choice(normal,size=5) for i in range(1000)]]

sns.histplot(mean5, ax= axes[2], binwidth=1,binrange=(mi-3*sig,mi+3*sig))

mean20 = [np.mean(lista) for lista in [ np.random.choice(normal,size=20) for i in range(1000)]]

sns.histplot(mean20, ax= axes[3], binwidth=1,binrange=(mi-3*sig,mi+3*sig))

mean75 = [np.mean(lista) for lista in [ np.random.choice(normal,size=75) for i in range(1000)]]

sns.histplot(mean75, ax= axes[4], binwidth=1,binrange=(mi-3*sig,mi+3*sig))

table = PrettyTable()

table.field_names = [
    'Var da Distri',
    'Var da Distri amstl medias-size=5',
    'Std da Distri',
    'Std da Distri amstl-médias-size=5'
]

table.add_row(
    [sig**2, 
     np.var(mean5)*5,
     sig,
     np.std(mean5)*np.sqrt(5)
     
    ]
)

table.add_row([
        'Var da Distri',
        'Var da Distri amstl médias-size=20',
        'Std da Distri',
        'Std da Distri amstl médias=size=20'
])

table.add_row(
    [sig**2, 
     np.var(mean20)*20,
     sig,
     np.std(mean20)*np.sqrt(20)
    ]
)

table.add_row([
        'Var da Distri',
        'Var da Distri amstl médias-size=75',
        'Std da Distri',
        'Std da Distri amstl médias=size=75'
])

table.add_row(
    [sig**2, 
     np.var(mean75)*75,
     sig,
     np.std(mean75)*np.sqrt(75)
    ]
)

table.hrules =True

print(table)
```

```python
setIntervals = []

for i in range(1000):
    scale = .2  # Inverso da taxa lambda (1/lambda)
    size = 1000  # Número de valores a serem gerados
    
    # Gerar valores
    amostra = np.random.exponential(scale, size)
    
    intervalo = ((size-(1.96*np.sqrt(size)))/np.sum(amostra), (size+(1.96*np.sqrt(size)))/np.sum(amostra))

    if intervalo[0]>5 or intervalo[1]<5:
        setIntervals.append(intervalo)

print(len(setIntervals))

print(len(setIntervals)*100/size)
```
