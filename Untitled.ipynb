```python
import numpy as np
from scipy import stats as stats

import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd

import os
```

```python
path = r'C:\Users\ander\Documents\practical-statistics-for-data-scientists\DataBase'

file = 'loans_income.csv'

filePath = os.path.join(path,file)
```

```python
dfLoan = pd.read_csv(filePath)

data = dfLoan['x']

sns.histplot(x=data, bins=np.histogram_bin_edges(a=data, bins=28), kde=True)

plt.show()
```

```python
shape, loc, scale = stats.lognorm.fit(data,floc=0)

x = np.linspace(data.min(), data.max(), 50000)

pdf = stats.lognorm.pdf(x, shape, loc, scale)

rvs = stats.lognorm.rvs(size=50000, s=shape, loc=loc, scale=scale)

fig, axes = plt.subplots(2, 1, figsize=(10, 6))
axes[0].hist(data, bins=28, density=True, alpha=0.5, color='blue', edgecolor='black',)
axes[0].plot(x, pdf, 'r-', lw=2, label='Lognormal PDF')
plt.legend()

sns.histplot(x=rvs, bins=np.histogram_bin_edges(a=data, bins=28), ax=axes[1] ) 

plt.show()
```

```python
stats.median_abs_deviation(rvs[rvs<200000]),stats.median_abs_deviation(data),
```
